```python
import json
import pandas as pd

import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
```

```python
auths_file = 'data/sn_auths.json'
users_age_file = 'data/sn_users_age.json'

with open(auths_file) as json_data:
    auths_data = json.load(json_data)
    
with open(users_age_file) as json_data:
    users_data = json.load(json_data)

auths_df = pd.DataFrame(auths_data)
users_age_df = pd.DataFrame(users_data)
```

```python
users_age_df
```

```python
fig = px.histogram(users_age_df, 
                   x='AGE',
                   category_orders=dict(AGE=['LESS THAN 2 DAYS',
                                             'LESS THAN 2 WEEKS',
                                             'LESS THAN 1 MONTH',
                                             'LESS THAN 1 QUARTER',
                                             'LESS THAN 1 YEAR',
                                             'MORE THAN 1 YEAR'
                                             ])
                  )

fig.update_layout(bargap=0.2,
                  title='Distribution of users by wallet age category when signing in Social.NEAR for the first time',
                  #x=0.5,
                  font=dict(color='black'
                           ),
                  hoverlabel=dict(bgcolor='white',
                                  font_size=12
                                 )
                 )

fig.update_xaxes(title='Wallet Age',
                 title_font=dict(size=14,
                                 color='black'
                                )
                )

fig.update_yaxes(title='Number of Wallets',
                 title_font=dict(size=14, 
                                 color='black'
                                )
                )

fig.update_traces(hovertemplate =
                  '<b>%{x}</b><br> <br>' +
                  '<i>Count</i>: ' + '%{y} wallets' + '<extra></extra>',
                  selector=dict(type="histogram"))


fig.show()
```

```python
fig = px.histogram(users_age_df, 
                   x='AGE_DAYS',
                   nbins=20)

fig.update_layout(bargap=0.1,
                  title='Distribution of users by wallet age in days when signing in Social.NEAR for the first time',
                  #x=0.5,
                  font=dict(color='black'
                           ),
                  hoverlabel=dict(bgcolor='white',
                                  font_size=12
                                 )
                 )

fig.update_xaxes(title='Wallet Age [days]',
                 title_font=dict(size=14,
                                 color='black'
                                )
                )

fig.update_yaxes(title='Number of Wallets',
                 title_font=dict(size=14, 
                                 color='black'
                                )
                )

fig.update_traces(hovertemplate =
                  '<i>Age</i>: ' + '%{x} days<br>' +
                  '<i>Count</i>: ' + '%{y} wallets' + '<extra></extra>',
                  selector=dict(type="histogram"))

fig.show()
```
